```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
# Đọc file Excel 
file_path = "D07 - parkinsons.data.xlsx"

# đọc tất cả sheet để kiểm tra
xls = pd.ExcelFile(file_path)
print(" Danh sách sheet:", xls.sheet_names)

df = pd.read_excel(file_path, sheet_name=xls.sheet_names[0])

# cột 'name' thì bỏ đi 
if 'name' in df.columns:
    df = df.drop(columns=['name'])

# Lưu file CSV
df.to_csv("data.csv", index=False)
print(" Đã lưu data.csv")
```

```python
print("5 dòng đầu tiên:")
display(df.head())

print("\nThông tin tổng quát:")
df.info()
```

```python
num_rows, num_cols = df.shape
print(f"Số hàng: {num_rows}")
print(f"Số cột: {num_cols}")
print(f"Số chiều: {df.ndim}")
print(f"Tổng số ô: {num_rows * num_cols}")
```

```python
print("Kiểu dữ liệu của từng cột:")
display(df.dtypes)
```

```python
if 'status' in df.columns:
    print("Số lượng mẫu theo nhãn status:")
    display(df['status'].value_counts())
else:
    print("Không tìm thấy cột 'status' trong dữ liệu!")
```

```python
# Chọn cột numeric
numeric_cols = df.select_dtypes(include=[np.number])

# Tính min / max / mean
stats = pd.DataFrame({
    'min': numeric_cols.min(),
    'max': numeric_cols.max(),
    'mean': numeric_cols.mean().round(3)
})

# Lưu ra file CSV
stats.to_csv("numeric_columns_stats.csv")
print(" Đã lưu numeric_columns_stats.csv")

display(stats.head())
```
